Step 1: Fetch and Extract Text from URLs

```python
import urllib.request
from html.parser import HTMLParser
```

```python
class SimpleHTMLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.text_parts = []

    def handle_data(self,data):
        self.text_parts.append(data)

    def get_text(self):
        return ' '.join(self.text_parts)
    

def fetch_text_from_url(url):
    try:
        with urllib.request.urlopen(url, timeout=10) as response:
            html = response.read().decode('utf-8', errors='ignore')
        
        parser = SimpleHTMLStripper()
        parser.feed(html)
        return parser.get_text().strip()
    
    except Exception as e:
        print(f"Error fetching {url}: {e}")
```

```python
import string

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('','', string.punctuation))

    tokens = text.split()
    # print(tokens)

    filtered_tokens = [word for word in tokens if word.isalpha() and word not in en_stop_words]

    return filtered_tokens
```

Step 3: Term Frequency (TF)

```python
from collections import Counter

def compute_tf(doc_tokens):
    term_counts = Counter(doc_tokens)
    total_terms = len(doc_tokens)
    tf = {term: count / total_terms for term, count in term_counts.items()}
    return tf
```

```python
tf_documents = [compute_tf(doc) for doc in preprocessed_docs]

for term, score in list(tf_documents[0].items())[:10]:
    print(f'{term}: {score:.4f}')
```

Step 4: Inverse Document Frequency (IDF)

```python
import math
from collections import defaultdict

def compute_idf(tf_documents):
    N = len(tf_documents)
    df = defaultdict(int)

    # Count how many documents contain each term
    for tf_doc in tf_documents:
        for term in tf_doc.keys():
            df[term] += 1

    # Compute IDF
    idf = {term: math.log(N / (1 + df[term])) for term in df}
    return idf

idf = compute_idf(tf_documents)

# Print IDF for first 10 terms
for term, score in list(idf.items())[:10]:
    print(f"{term}: {score:.4f}")
```

Step 5: Compute TF-IDF Vectors

```python
def compute_tfidf(tf_documents, idf):
    tfidf_documents = []

    for tf_doc in tf_documents:
        tfidf = {}
        for term, tf_value in tf_doc.items():
            tfidf[term] = tf_value * idf.get(term, 0.0)
        tfidf_documents.append(tfidf)

    return tfidf_documents

tfidf_documents = compute_tfidf(tf_documents, idf)

# Print TF-IDF values for first document
for term, score in list(tfidf_documents[0].items())[:10]:
    print(f"{term}: {score:.4f}")
```

Step 6: User Query → Search & Rank Using Cosine Similarity

```python
import math

def cosine_similarity(vec1, vec2):
    # Set of all unique terms
    all_terms = set(vec1.keys()).union(vec2.keys())

    # Dot product and norms
    dot_product = sum(vec1.get(term, 0.0) * vec2.get(term, 0.0) for term in all_terms)
    norm1 = math.sqrt(sum((vec1.get(term, 0.0))**2 for term in all_terms))
    norm2 = math.sqrt(sum((vec2.get(term, 0.0))**2 for term in all_terms))

    if norm1 == 0 or norm2 == 0:
        return 0.0
    return dot_product / (norm1 * norm2)

def search(query, tfidf_documents, idf):
    query_tokens = preprocess(query)
    query_tf = compute_tf(query_tokens)
    query_tfidf = {term: query_tf[term] * idf.get(term, 0.0) for term in query_tf}

    scores = []
    for i, doc_vector in enumerate(tfidf_documents):
        sim = cosine_similarity(query_tfidf, doc_vector)
        scores.append((i, sim))

    ranked = sorted(scores, key=lambda x: x[1], reverse=True)
    return ranked
```

```python
query = "machine learning and natural language"
results = search(query, tfidf_documents, idf)

for idx, score in results[:3]:  # Show top 3
    print(f"Document {idx} (Score: {score:.4f}):")
    print(documents[idx][:200], "\n---\n")
```
